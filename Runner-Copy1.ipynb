```python
#Create Temporal network
nlist = [25]
p = 0.0002
t = 15000
for n in nlist:
    name = "data/ER/ER_n"+str(n)+"_p"+str(p)[2:]+"_T"+str(t) +".edges"
    A = [nx.gnp_random_graph(n, p, seed=None, directed=False) for i in range(t)]
    if len(A[0].edges()) ==0:
        u,v = np.random.choice(range(n),2,replace = False )
        A[0].add_edge(u,v)
    if len(A[-1].edges()) ==0:
        u,v = np.random.choice(range(n),2,replace = False )
        A[-1].add_edge(u,v)
    SubTempNet.write_snapshotlist_to_edgelist(A, name, separator='\t')
    filepath ="data/ER/ER_n"+str(n)+"_p0002"+"_T15000.edges"
    objname = "ER_n"+str(n)+"_p0002"
    STN= SubTempNet(filepath,objname,directed= False, init = True)
    STN.save()
```
