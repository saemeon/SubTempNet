```python
#names=["primaryschool","hospital","highschool2012","lotr","hypertext","SFHH"]
names=["primaryschool","hospital","hypertext"]
degseq=[]
dh=[]
d = []
for name in names:
    filename = "data/"+name+".edges"
    objname= name
    STN = SubTempNet(filename, objname, directed= False, init = False)
    print(STN)
    W=SubTempNet.aggregate_Matrices(STN["A"], weighted = True)
    AW = nx.from_scipy_sparse_matrix(W,parallel_edges=True,create_using=nx.MultiGraph)
    selfloops = nx.selfloop_edges(AW.copy())
    AW.remove_edges_from(selfloops)
    deg_hist = nx.degree_histogram(AW)
    degseq.append(STN["deg_seq"])
    #dh.append(nx.degree_histogram(AW))
    #d.append(range(len(nx.degree_histogram(AW))))
    dh.append(np.array(nx.degree_histogram(AW))/nx.number_of_nodes(AW))
    print(len(np.array(nx.degree_histogram(AW))/nx.number_of_nodes(AW)))
```
